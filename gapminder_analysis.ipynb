```python
def pop_year(df, year):
    '''Function that calculates total population per continent for a year. 
    Input: dataframe, year, Output: resulting dataframe'''#dot string will show when function queried using ?
    #with checks to make sure that inputs are possible
    assert type(year) is int, "Year is not an integer" #if assert fails, show message
    pop_data = df[(df.year == year)][["continent","pop"]].groupby("continent").sum().sort("pop") #You can pipe a whole bunch of functions!
    return pop_data
```

Make a plotting function

```python
def pop_plot(df,year):
    year_string = str(year)
    types = list(df.index)
    y_pos = np.arange(len(types))

    plt.barh(y_pos, df["pop"], align='center')
    plt.yticks(y_pos, types)
    plt.xlabel('total population size')
    plt.ylabel('continent')
    plt.title('population of continents '+year_string)
    plt.show()
```
